```python
# Linear Regression
# Task -In this regression task we will predict the percentage of marks that a student is expected to           score based upon the number of hours they studied. This is a simple linear regression task as i         it involves just two variables.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reading Data from csv file
df = pd.read_csv('data.csv')
df
```

```python
# Plot Graph using data points to find relation between them.
plt.style.use('fivethirtyeight')
plt.grid(True)
plt.title("Hours Vs Percentage ")
plt.scatter(x = df.Hours,y = df.Scores);
plt.xlabel("Hours Studied");
plt.ylabel("Percentage Score");
plt.legend();
```

```python
# Divide data in two parts inputs and outputs
X = df.iloc[:, :-1].values  
y = df.iloc[:, 1].values 
```

```python
# Split data into training sets and test sets.
from sklearn.model_selection import train_test_split 
X_train,X_test,Y_train,Y_test = train_test_split(X,y,test_size=0.2, random_state=0)
```

```python
# We have split our data into training and testing sets, and now is finally the time to train our algorithm.
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train,Y_train)
print("Training Complete")
```

```python
# Plot regression line
line = reg.coef_*X+reg.intercept_
plt.title("Regression Line")
plt.scatter(X,y)
plt.plot(X,line,'r')
plt.xlabel('Hours');
plt.ylabel("Scores");
plt.legend();
```

```python
# We trained our algorithm now make some predictions
y_pred = reg.predict(X_test)
```

```python
# Comparing Actual Vs Predicted 
new_df = pd.DataFrame({'Actual':Y_test,'Predicted':y_pred})
new_df
```

```python
# predicted score if a student studies for 9.25 hrs/ day
no_hours = 9.25
predicted_score = reg.predict([[no_hours]])
print(f"No of Hours Student Studied : {no_hours}")
print(f"Predicted Score for {no_hours} is {predicted_score[0]}")
```

```python
# Evaluating the model
# The final step is to evaluate the performance of algorithm. This step is particularly important to        compare how well different algorithms perform on a particular dataset. For simplicity here, we          have chosen the mean square error.
from sklearn import metrics  
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(Y_test, y_pred)) 
```
